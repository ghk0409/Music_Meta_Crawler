```python
from Module.album_list_crawling import album_list_crawler
from Module.meta_album_crawling import album_crawler
from Module.meta_artist_crawling import artist_crawler
from Module.meta_music_crawling import music_crawler
from Module.selenium_driver import webdriver_control
from Module.utility import result2csv
import sys
import pandas as pd
import time
from tqdm.notebook import tqdm_notebook
```

```python
# 국내 장르별 리스트
nation_korea = [
    'ballad', 'dance', 'hiphop', 'rnb', 'indie', 'rock', 'trot', 'folk',
]
# 해외 장르별 리스트
nation_overseas = [
    'pop', 'rock', 'elec', 'hiphop', 'rnb', 'folk',
]
# 그외 장르별 리스트
nation_etc = [
    'ost', 'classic', 'jazz', 'newage', 'jpop',  'world', 'ccm', 'child1',
    'child2', 'child3', 'child4', 'child5', 'child6', 'child7', 'child8',
    'child9', 'child10', 'child11', 'catholic', 'buddhism', 'traditional', 
]
```

```python
# 먼저 cmd에서 크롬창 띄워줘야 자동화탭 안 뜸
# cd C:/Program Files/Google/Chrome/Application
# chrome.exe --remote-debugging-port=9222 --user-data-dir=C:/Chrome_debug_temp

# 드라이버 경로
DRIVER_PATH = 'D:/'

# 운영체제별 드라이버 선택
platform = sys.platform
# 윈도우일 경우
if platform == 'win32':
    DRIVER_PATH += 'chromedriver.exe'
# 리눅스일 경우
elif platform == 'linux':
    DRIVER_PATH += 'chromedriver_linux.exe'

# Chrome Driver init
chrome = webdriver_control.generate_chrome(driver_path=DRIVER_PATH)
```

```python
genre_name = nation_korea[0]
alc = album_list_crawler(chrome, 0, genre_name, page=30)
new_album_idx = alc.extract_album_list()
```

```python
for i in tqdm_notebook(range(len(new_album_idx))): # len(new_album_idx)
    # 앨범 메타 데이터 추출
    try:
        abc = album_crawler(chrome, new_album_idx[i])
        time.sleep(0.5)
        df_album, album_artists, album_musics = abc.return_result()
        result2csv(f'./csv/new_albums_{genre_name}.csv', df_album)
    except:
        continue
    
    # 아티스트 메타 데이터 추출
    for idx in album_artists:
        if idx == 0:
            pass
        else:
            try:
                atc = artist_crawler(chrome, idx)
                time.sleep(0.5)
                df_artist = atc.return_result()
                result2csv(f'./csv/new_artists_{genre_name}.csv', df_artist)
            except:
                continue
            
    # 음원 메타 데이터 추출
    for idx in album_musics:
        try:
            msc = music_crawler(chrome, idx, new_album_idx[i])
            time.sleep(0.5)
            df_music = msc.return_result()
            result2csv(f'./csv/new_musics_{genre_name}.csv', df_music)
        except:
            continue
```
